```python
import numpy as np
import pandas as pd
from abc import ABC, abstractmethod
from dataclasses import dataclass
from typing import Callable, List, Optional
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
```

# Signal Generator

```python
class SignalGenerator(ABC):
    @abstractmethod
    def generate_signals(self, data: pd.DataFrame) -> pd.Series:
        """
        data: OHLCV (and extra) up to t‑1
        returns: Series indexed like data of +1 (long), -1 (short), 0 (flat)
        """
```

```python
class SimpleMAMomentum(SignalGenerator):
    """Long when price > its 24h SMA, exit when below 24h SMA"""
    def __init__(self, window: int = 24):
        self.window = window

    def generate_signals(self, data: pd.DataFrame) -> pd.Series:
        close = data['close']
        sma = close.rolling(self.window).mean()
        signal = pd.Series(0, index=data.index)
        signal[close > sma] = 1
        signal[close < sma] = -1
        return signal
```

```python
class EnhancedMACDRSIStrategy(SignalGenerator):
    """Refined MACD+RSI strategy: includes trend filtering and volatility suppression"""
    def __init__(self,
                 macd_fast:int=12, macd_slow:int=26, macd_signal:int=9,
                 rsi_period:int=14, rsi_oversold:float=30, rsi_overbought:float=70,
                 trend_ema:int=100, min_volatility:float=0.005):
        self.fast = macd_fast
        self.slow = macd_slow
        self.signal = macd_signal
        self.rsi_period = rsi_period
        self.rsi_over = rsi_overbought
        self.rsi_under = rsi_oversold
        self.trend_ema = trend_ema
        self.min_vol = min_volatility

    def generate_signals(self, data:pd.DataFrame) -> pd.Series:
        df = data.copy()
        close = df['close']
        # MACD: fast EMA, slow EMA
        ema_fast = close.ewm(span=self.fast, adjust=False).mean()
        ema_slow = close.ewm(span=self.slow, adjust=False).mean()
        macd_line = ema_fast - ema_slow
        macd_signal = macd_line.ewm(span=self.signal, adjust=False).mean()
        # RSI: standard Wilder's RSI
        delta = close.diff()
        up = delta.clip(lower=0)
        down = -delta.clip(upper=0)
        ema_up = up.ewm(alpha=1/self.rsi_period, adjust=False).mean()
        ema_down = down.ewm(alpha=1/self.rsi_period, adjust=False).mean()
        rs = ema_up / ema_down
        rsi = 100 - (100 / (1 + rs))

                sig = pd.Series(0, index=df.index)

        # Volatility filter (relative daily range)
        atr = (df['high'] - df['low']) / df['close'].shift(1)
        low_vol = atr < self.min_vol

        # Trend filter (price must be above long-term EMA)
        trend = df['close'] > df['close'].ewm(span=self.trend_ema, adjust=False).mean()

        # bullish MACD crossover + RSI oversold + volatility/trend filter
        buy = (
            (macd_line > macd_signal) &
            (macd_line.shift() <= macd_signal.shift()) &
            (rsi < self.rsi_under) &
            trend &
            (~low_vol)
        )
        sell = (
            (macd_line < macd_signal) &
            (macd_line.shift() >= macd_signal.shift()) &
            (rsi > self.rsi_over)
        )

        sig[buy] = 1
        sig[sell] = -1
        return sig
```

# Backtesting Framework

```python
@dataclass
class BacktestConfig:
    fee: float = 0.0006                   # 0.06% commission
    spread: float = 0.0002                # 2bp half‑spread
    slippage_fn: Callable[[float, float], float] = lambda price, qty: 0.0
    start_equity: float = 10_000.0
    risk_per_trade: float = 0.01           # 1% of equity
    leverage: float = 1.0
    funding_rate: float = 0.0001           # per hour
    stop_loss: float = 0.02                # 2%
    take_profit: float = 0.05              # 5%
    margin_requirement: float = 0.5        # maintenance margin
    min_qty: float = 0.001
    tick_size: float = 0.01
    freq_hours: float = 1                  # bar frequency in hours

    @property
    def annual_factor(self) -> float:
        periods = 365 * (24 / self.freq_hours)
        return np.sqrt(periods)
```

```python
class CostModel:
    def __init__(self, cfg: BacktestConfig):
        self.cfg = cfg

    def cost_pct(self, price: float, qty: float) -> float:
        return (
            self.cfg.fee +
            2 * self.cfg.spread +
            self.cfg.slippage_fn(price, qty)
        )
```

```python
class PositionManager:
    def __init__(self, cost_model: CostModel, cfg: BacktestConfig):
        self.cost_model = cost_model
        self.cfg = cfg
        self.position: Optional[Position] = None
        self.trades: List[Position] = []

    def on_bar(self,
               time: pd.Timestamp,
               bar: pd.Series,
               signal: int,
               equity: float) -> float:
        price_o, price_h, price_l = bar['open'], bar['high'], bar['low']
        realized_pnl = 0.0

        # ENTRY
        if self.position is None and signal != 0:
            raw_qty = equity * self.cfg.risk_per_trade / (self.cfg.stop_loss * price_o)
            lev_qty = raw_qty * self.cfg.leverage
            qty = np.floor(lev_qty / self.cfg.min_qty) * self.cfg.min_qty
            side = int(np.sign(signal))
            cost = self.cost_model.cost_pct(price_o, qty)
            entry_price = price_o * (1 + side * cost)
            self.position = Position(time, entry_price, qty, side)
            return 0.0

        # EXIT / SL / TP / MARGIN-CALL
        if self.position:
            pos = self.position
            sl_level = pos.entry_price * (1 - pos.side * self.cfg.stop_loss)
            tp_level = pos.entry_price * (1 + pos.side * self.cfg.take_profit)
            exit_price = None

            # 1) Stop-loss
            if (pos.side > 0 and price_l <= sl_level) or (pos.side < 0 and price_h >= sl_level):
                cost = self.cost_model.cost_pct(sl_level, pos.qty)
                exit_price = sl_level * (1 - pos.side * cost)
            # 2) Take-profit
            elif (pos.side > 0 and price_h >= tp_level) or (pos.side < 0 and price_l <= tp_level):
                cost = self.cost_model.cost_pct(tp_level, pos.qty)
                exit_price = tp_level * (1 - pos.side * cost)
            # 3) Signal-driven exit
            elif signal == 0 or np.sign(signal) != pos.side:
                cost = self.cost_model.cost_pct(price_o, pos.qty)
                exit_price = price_o * (1 - pos.side * cost)
            # 4) Margin call
            if exit_price is None:
                margin_req = pos.notional() * self.cfg.margin_requirement
                if equity < margin_req:
                    worst = price_l if pos.side > 0 else price_h
                    cost = self.cost_model.cost_pct(worst, pos.qty)
                    exit_price = worst * (1 - pos.side * cost)

            if exit_price is not None:
                realized_pnl = pos.side * pos.qty * (exit_price - pos.entry_price)
                pos.exit_time  = time
                pos.exit_price = exit_price
                pos.pnl        = realized_pnl
                self.trades.append(pos)
                self.position = None

        return realized_pnl
```

```python
class MetricsTracker:
    def __init__(self, cfg: BacktestConfig):
        self.cfg = cfg
        self.times: List[pd.Timestamp] = []
        self.equity_curve: List[float] = []

    def record(self, time: pd.Timestamp, equity: float):
        self.times.append(time)
        self.equity_curve.append(equity)

    def compute(self, trades: List[Position]) -> Dict[str, float]:
        eq = pd.Series(self.equity_curve, index=self.times)
        returns = eq.pct_change().dropna()
        ann = self.cfg.annual_factor

        sharpe = returns.mean() / returns.std() * ann if returns.std() else np.nan
        cum = (1 + returns).cumprod()
        dd = (cum - cum.cummax()) / cum.cummax()
        max_dd = dd.min()

        wins = [t.pnl for t in trades if t.pnl > 0]
        losses = [t.pnl for t in trades if t.pnl < 0]
        win_rate = len(wins) / len(trades) if trades else 0.0
        avg_win = np.mean(wins) if wins else 0.0
        avg_loss = np.mean(losses) if losses else 0.0
        profit_factor = sum(wins) / abs(sum(losses)) if losses else np.inf
        freq = len(trades) / len(eq) if len(eq) else 0.0

        return {
            'Sharpe':        sharpe,
            'Max Drawdown':  max_dd,
            'Win Rate':      win_rate,
            'Avg Win':       avg_win,
            'Avg Loss':      avg_loss,
            'Profit Factor': profit_factor,
            'Trade Freq':    freq,
            'Total Trades':  len(trades)
        }
```

```python
class Backtester:
    def __init__(self,
                 data: pd.DataFrame,
                 strategy: Callable[[pd.DataFrame], pd.Series],
                 cfg: BacktestConfig):
        base = data.sort_index()
        idx = pd.date_range(base.index.min(), base.index.max(), freq='1h')
        self.data = base.reindex(idx).ffill().dropna(subset=['open','high','low','close','volume'])
        self.strategy = strategy
        self.cfg = cfg

    def run(self) -> Dict[str, float]:
        cost_model = CostModel(self.cfg)
        pm = PositionManager(cost_model, self.cfg)
        mt = MetricsTracker(self.cfg)

        # initialize equity with progress bar
        mt.record(self.data.index[0], self.cfg.start_equity)
        cash = self.cfg.start_equity

        for t in tqdm(range(1, len(self.data)), desc="Backtesting", unit="bar"):
            time = self.data.index[t]
            bar = self.data.iloc[t]
            past = self.data.iloc[:t]
            signal = self.strategy(past).iloc[-1]

            if pm.position:
                cash -= abs(pm.position.qty) * bar['close'] * self.cfg.funding_rate

            unreal = pm.position.side * pm.position.qty * (bar['close'] - pm.position.entry_price) \
                        if pm.position else 0.0
            equity_before = cash + unreal

            pnl = pm.on_bar(time, bar, int(signal), equity_before)
            cash += pnl

            unreal_after = pm.position.side * pm.position.qty * (bar['close'] - pm.position.entry_price) \
                          if pm.position else 0.0
            equity_after = cash + unreal_after
            mt.record(time, equity_after)

        metrics = mt.compute(pm.trades)
        self._plot_equity(mt)
        self._plot_drawdown(mt)
        self._plot_rolling(mt)
        self._plot_returns_dist(mt)
        self._plot_underwater(mt)
        self._plot_trade_scatter(pm)
        return metrics

    # ───────── PLOTTING ──────────────────────────────────────────────────────
    def _plot_equity(self, mt: MetricsTracker):
        eq = pd.Series(mt.equity_curve, index=mt.times)
        plt.figure(); eq.plot(); plt.title("Equity Curve"); plt.show()

    def _plot_drawdown(self, mt: MetricsTracker):
        eq = pd.Series(mt.equity_curve, index=mt.times)
        dd = (1+eq.pct_change().fillna(0)).cumprod()
        dd = (dd - dd.cummax())/dd.cummax()
        plt.figure(); dd.plot(); plt.title("Drawdown Curve"); plt.show()

    def _plot_rolling(self, mt: MetricsTracker, window:int=24):
        eq = pd.Series(mt.equity_curve, index=mt.times)
        ret = eq.pct_change().fillna(0)
        ann = self.cfg.annual_factor
        rs = (ret.rolling(window).mean()/ret.rolling(window).std())*ann
        rv = ret.rolling(window).std()*ann
        plt.figure(); rs.plot(label="Rolling Sharpe"); rv.plot(label="Rolling Volatility"); plt.legend(); plt.show()

    def _plot_returns_dist(self, mt: MetricsTracker):
        eq = pd.Series(mt.equity_curve, index=mt.times)
        ret = eq.pct_change().dropna()
        plt.figure(); ret.hist(bins=50, density=True); plt.title("Returns Distribution"); plt.show()

    def _plot_underwater(self, mt: MetricsTracker):
        eq = pd.Series(mt.equity_curve, index=mt.times)
        cum = (1+eq.pct_change().fillna(0)).cumprod()
        dd = (cum - cum.cummax())/cum.cummax()
        plt.figure(); plt.fill_between(dd.index, dd.values, 0); plt.title("Underwater Plot"); plt.show()

    def _plot_trade_scatter(self, pm: PositionManager):
        prices = self.data['close']
        plt.figure(); prices.plot(label="Price")
        entries = [(t.entry_time, t.entry_price) for t in pm.trades]
        exits   = [(t.exit_time,  t.exit_price)  for t in pm.trades]
        if entries:
            e_t,e_p = zip(*entries); plt.scatter(e_t, e_p, marker='^', c='g', label='Entry')
            x_t,x_p = zip(*exits); plt.scatter(x_t, x_p, marker='v', c='r', label='Exit')
        plt.legend(); plt.title("Trade Scatter"); plt.show()
```
